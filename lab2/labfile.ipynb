```python
# Question 7: Find key with maximum value without using max()

products = {
    'pen': 10,
    'bag': 300,
    'calculator': 200,
    'laptop': 1500,
    'mouse': 50
}

max_key = None
max_value = float('-inf') # Start with -infinity

for key, value in products.items():
    if value > max_value:
        max_value = value
        max_key = key

print(f"Product with highest price: {max_key} ==>{max_value}")
```

```python
class EvenIterator:
    def __init__(self, numbers):
        self.numbers = numbers
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        while self.index < len(self.numbers):
            num = self.numbers[self.index]
            self.index += 1

            if num % 2 == 0:
                return num

        raise StopIteration


nums = [1, 2, 3, 4, 5, 6]

even_numbers = EvenIterator(nums)

for i in even_numbers:
    print(i)
```

```python
class SquareIterator:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.limit:
            result = self.current ** 2
            self.current += 1
            return result
        else:
            raise StopIteration

sq = SquareIterator(5)

for i in sq:
    print(i)
```

# Generators


```python
def numbers(n):
    for i in range(1, n + 1):
        yield i

for x in numbers(5):
    print(x)
```

```python
def gen_square(n):
    for i in range(1, n+1):
        yield i*i

for s in gen_square(5):
    print(s)

```

```python
def gen_fibo(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a+b

for f in gen_fibo(5):
    print(f)

```

```python
def gen_prime(a, b):
    for num in range(a, b+1):
        if num > 1:
            for i in range(2, num):
                if num % i == 0:
                    break
            else:
                yield num

for p in gen_prime(10, 20):
    print(p)
```

## decorator

```python
def deco(func):
    def wrap():
        print("Before")
        func()
        print("After")
    return wrap

@deco
def show():
    print("Hello")

show()
```

```python
def count(func):
    c = 0
    def wrap():
        nonlocal c
        c += 1
        print("Called", c, "times")
        func()
    return wrap

@count
def hi():
    print("Hi")

hi()
hi()
```

```python
def upper(func):
    def wrap():
        return func().upper()
    return wrap

@upper
def msg():
    return "hello"

print(msg())
```
