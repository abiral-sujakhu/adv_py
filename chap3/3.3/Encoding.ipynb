```python
# List of reasons to encode categorical data:

# Machine learning algorithms require numerical input

# Enable mathematical operations

# Handle non-numeric relationships

# Improve model performance

# Reduce memory usage for high-cardinality features

# Facilitate feature engineering

# Ensure compatibility with pipelines
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Blue', 'Green', 'Blue']
le = LabelEncoder()
encoded = le.fit_transform(data)
encoded
```

```python
#One-Hot Encoding
import pandas as pd

data = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Blue']})
one_hot = pd.get_dummies(data, columns=['Color'])
one_hot
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

data = [['Low'], ['Medium'], ['High'], ['Medium']]
encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
encoded = encoder.fit_transform(data)
encoded
```

```python
#Frequency Encoding
data = pd.DataFrame({'City': ['Paris', 'London', 'Berlin', 'Paris']})
freq = data['City'].value_counts()
data['City_freq'] = data['City'].map(freq)
print(data)
```
