1Ô∏è‚É£ Collections (from collections module)

üëâ Meaning (simple):
Collections are special versions of normal data types (like list, dict, tuple) that make some tasks easier and faster.

````python
# Named tuple =L ike a tuple, but with names for values

from collections import namedtuple

Student = namedtuple("Student", ["name", "age"])
s1 = Student("Ram", 20)

print(s1.name)   # Ram
print(s1.age)    # 20
````

````python
# deque (Double Ended Queue)
#  üëâ A list where you can add/remove items from both ends easily

from collections import deque

d = deque([1, 2, 3])
d.append(4)        # add to right
d.appendleft(0)    # add to left

print(d)
````

````python
# Counter = Counts how many times each item appears

from collections import Counter

nums = [1,1, 2, 2,2,2,2,2,3, 3, 3,3,3]
c = Counter(nums)
print(c)
````

````python
# defaultdict= A dictionary that never gives KeyError 
# Automatically gives default value (0 here).

from collections import defaultdict

d = defaultdict(int)
d["a"] += 1
d["b"] += 2

print(d)
````

2Ô∏è‚É£ Iterators

üëâ Meaning (simple):
An iterator is an object that gives one value at a time, instead of all at once. USES iter() and next()

````python
nums = [10, 20, 30]
it = iter(nums)

print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
````

3Ô∏è‚É£ Generators

üëâ Meaning (simple):
Generators produce values (lazily) one at a time using yield, saving memory.

```Lazily means: values are created only when they are needed, not all at once. ```

````python
def count_up(n):
    for i in range(1, n+1):
        yield i   #return ‚Üí stops function
                   #yield ‚Üí pauses and continues later

g = count_up(3)

print(next(g))  # 1
print(next(g))  # 2
print(next(g))  # 3
````

4Ô∏è‚É£ Decorators

üëâ Meaning (simple):
A decorator is a function that adds extra features to another function without changing its code.

Real-life idea:

üéÅ Gift wrapping = decorator

üéÅ Gift inside = original function

````python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
````
