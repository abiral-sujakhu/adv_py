```python
import pandas as pd
import json

# For nested JSON, read as dictionary first then normalize
with open("data.json", "r") as file:
    data = json.load(file)
    
# Convert to DataFrame using json_normalize for nested data
df3 = pd.json_normalize(data)
print(df3)
```
