## default constructor

```python
class Myclass:
    def __init__(self): #constructor
        print("hellow world")
m1=Myclass()
```

### parameterized constructor

```python
class Myclass:
    def __init__(self,name,age):  # self is the default argument 
        self.name=name
        self.age=age
    def display(self):
        print(self.name)
        print(self.age)
m1=Myclass("aviral",20)
m1.display()
m2=Myclass("ram",22)
m2.display()

```

## inheritance

```python
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print(f"{self.name} make sound")
class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks")
a=Animal("generic animal")
b=Dog("buddy")
a.speak()
b.speak()
```

wap to illustrate multiple inheritance

```python
class living_being:
    def __init__(self,category):
        self.category=category

    def divison(self):
        print(f"{self.category} type")

class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print(f"{self.name} make sound")

class Dog(living_being,Animal):
    def __init__(self, name, category):
        living_being.__init__(self,category)
        Animal.__init__(self,name)
    def divison(self):
        print(f"{self.category} group")

    def speak(self):
        print(f"{self.name} barks")

c=living_being("biological")
a=Animal("generic animal")
b=Dog("buddy","mammal")



c.divison()
a.speak()

b.divison()
b.speak()

```
